## Fuerza Sol sobre la Tierra

````python
# Constantes y par√°metros
G = pc.constantes.G
m_tierra = pc.constantes.M_earth
m_sol = pc.constantes.M_sun
mu_sol = G * m_sol

# Posiciones del Sol y la Tierra respecto al SSB
# Tierra
tabla_T, jd_T, X_T = pc.consulta_horizons(id='399', location='@SSB', epochs='2026-02-19')
r_T = X_T[0:3]

# Sol
tabla_S, jd_S, X_S = pc.consulta_horizons(id='10', location='@SSB', epochs='2026-02-19')
r_S = X_S[0:3]
````



>Nota t√©cnica: en Horizons el Sol es id='10'


````python
# Vector relativo Sol-Tierra
r_ST = r_T - r_S
distancia = np.linalg.norm(r_ST)

# Fuerza Gravitacional
F_ST = -G * m_sol * m_tierra * r_ST / distancia**3
````

````python
F_mag = np.linalg.norm(F_ST)

print("Vector posici√≥n Sol ‚Üí Tierra (km):")
print(r_ST)

print("\nDistancia Sol-Tierra (km):")
print(distancia)

print("\nMagnitud de la fuerza (N):")
print(F_mag)

print("\nVector fuerza del Sol sobre la Tierra (N):")
print(F_ST)
````

## Fuerza del Sol sobre la **Luna**

````python
tabla_L, jd_L, X_L = pc.consulta_horizons(id='301', location=location, epochs=epoch)
tabla_S, jd_S, X_S = pc.consulta_horizons(id='10', location=location, epochs=epoch)

r_L = X_L[0:3]
r_S = X_S[0:3]
````

Primero, definamos una funci√≥n general para 2 cuerpos:

````python
def fuerza_gravitacional(m1, m2, r1, r2):
    r = r2 - r1
    d = np.linalg.norm(r)
    return -G * m1 * m2 * r / d**3
````

````python
F_ST = fuerza_gravitacional(m_sol, m_tierra, r_S, r_T)
print('Vector fuerza del Sol sobre la Tierra (N):')
print(F_ST)
````

# ¬°Fuerzas en el sistema de Sol - Tierra - Luna!

üåû üåù üåé

Es posible aplicar los m√©todos previos en sentido inverso, para estudiar las fuerzas que son ejercidas sobre el Sol. Pero qu√© tosco y aburrido... para hacerlo mucho m√°s interesante:

 *¬°estudiemos ya el sistema de los 3 cuerpos y las influencias netas sobre cada uno!*

````python
mu_T, mu_L = G * m_tierra , G * m_luna # Ya hab√≠amos definido mu_sol

# Vectores relativos
r_ST = r_S - r_T
r_SL = r_S - r_L
r_TS = -r_ST
r_TL = r_T - r_L
r_LS = -r_SL
r_LT = -r_TL
````

### ‚òÄ Aceleraci√≥n del Sol debida a la Tierra y la Luna ‚òÄ

````python
# Tomamos fuerzas debidas tanto a Tierra como a Luna
a_S = (
    - mu_T * r_ST / np.linalg.norm(r_ST)**3
    - mu_L * r_SL / np.linalg.norm(r_SL)**3
)
````

### Funci√≥n general para N cuerpos

````python
def aceleraciones_n_cuerpos(r, mu):
  N = len(r)
  a = np.zeros_like(r)

  for i in range(N):
      for j in range(N):
          if i != j:
              rij = r[i] - r[j]
              a[i] -= mu[j] * rij / np.linalg.norm(rij)**3

  return a
````

Para preparar los datos de la funci√≥n `aceleraciones_n_cuerpos`, crearemos dos listas: una para los vectores de posici√≥n y otra para los par√°metros gravitacionales, siguiendo el orden especificado de Sol, Tierra y Luna.


````python
r = [r_S, r_T, r_L]
mu = [mu_sol, mu_T, mu_L]

print("Lista de vectores de posici√≥n (Sol, Tierra, Luna):")
for i, pos in enumerate(r):
    print(f"  Cuerpo {i+1}: {pos}")

print("\nLista de par√°metros gravitacionales (Sol, Tierra, Luna):")
for i, p_grav in enumerate(mu):
    print(f"  Cuerpo {i+1}: {p_grav}")
````

### C√°lculo de aceleraciones netas

Utilicemos la funci√≥n `aceleraciones_n_cuerpos` para calcular la aceleraci√≥n neta de cada uno de los tres cuerpos (Sol, Tierra, Luna) debido a la influencia gravitacional de los otros dos.


````python
a_netas = aceleraciones_n_cuerpos(r, mu)

print("Aceleraciones netas (Sol, Tierra, Luna) (km/s^2):")
for i, acc in enumerate(a_netas):
    print(f"  Cuerpo {i+1}: {acc}")
````

## C√°lculo de fuerzas netas

Multiplicaremos las aceleraciones netas de cada cuerpo por su respectiva masa para obtener las fuerzas gravitacionales netas actuando sobre el Sol, la Tierra y la Luna.


Definamos una lista de masas para el Sol, la Tierra y la Luna en el orden correcto para que se corresponda con la matriz `a_netas`. Luego, iteraremos sobre las masas y las aceleraciones netas para calcular las fuerzas netas y almacenarlas en `F_netas`.


````python
masses = [m_sol, m_tierra, m_luna]

F_netas = []
for i in range(len(masses)):
    F_netas.append(masses[i] * a_netas[i])

print("Fuerzas netas (Sol, Tierra, Luna) (N):")
body_names = ["Sol", "Tierra", "Luna"]
for i, force in enumerate(F_netas):
    print(f"  Fuerza neta sobre el {body_names[i]}: {force}")
````

````python
import plotly.graph_objects as go

# Extraer coordenadas x, y, z para cada cuerpo
x_S, y_S, z_S = r_S
x_T, y_T, z_T = r_T
x_L, y_L, z_L = r_L

# Crear la figura
fig = go.Figure()

# A√±adir el Sol
fig.add_trace(go.Scatter3d(
    x=[x_S], y=[y_S], z=[z_S],
    mode='markers',
    marker=dict(size=10, color='gold'),
    name='Sol'
))

# A√±adir la Tierra
fig.add_trace(go.Scatter3d(
    x=[x_T], y=[y_T], z=[z_T],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Tierra'
))

# A√±adir la Luna
fig.add_trace(go.Scatter3d(
    x=[x_L], y=[y_L], z=[z_L],
    mode='markers',
    marker=dict(size=3, color='silver'),
    name='Luna'
))

# Conectar Sol-Tierra
fig.add_trace(go.Scatter3d(
    x=[x_S, x_T], y=[y_S, y_T], z=[z_S, z_T],
    mode='lines',
    line=dict(color='gray', width=2, dash='dash'),
    name='Sol-Tierra'
))

# Conectar Tierra-Luna
fig.add_trace(go.Scatter3d(
    x=[x_T, x_L], y=[y_T, y_L], z=[z_T, z_L],
    mode='lines',
    line=dict(color='lightgray', width=1, dash='dot'),
    name='Tierra-Luna'
))

# Actualizar layout de la figura para fondo negro y texto blanco
fig.update_layout(
    title=dict(
        text=f'Disposici√≥n del Sistema Sol-Tierra-Luna el {epoch} (respecto al SSB)',
        font=dict(color='white') # T√≠tulo en blanco
    ),
    scene=dict(
        xaxis=dict(title='X (km)',  backgroundcolor="black", gridcolor="gray", zerolinecolor="gray", showbackground=True, tickfont=dict(color='white'), titlefont=dict(color='white')),
        yaxis=dict(title='Y (km)',  backgroundcolor="black", gridcolor="gray", zerolinecolor="gray", showbackground=True, tickfont=dict(color='white'), titlefont=dict(color='white')),
        zaxis=dict(title='Z (km)',  backgroundcolor="black", gridcolor="gray", zerolinecolor="gray", showbackground=True, tickfont=dict(color='white'), titlefont=dict(color='white')),
        aspectmode='data' # Mantener la proporci√≥n real de los ejes
    ),
    paper_bgcolor='black', # Fondo del papel en negro
    plot_bgcolor='black'   # Fondo del gr√°fico en negro
)

fig.show()
````
